```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

#### Problema 1

```python
def canales(imagen, color):
    #Inversión de colores a RGB
    img = cv.imread(imagen, cv.COLOR_BGR2RGB)
    R, G, B = cv.split(img) # Separación de canales
    imagen = np.zeros_like(img)

    # Seleccionamos el canal deseado
    if color == 1:
        imagen[:,:,2] = B
    elif color == 2:
        imagen[:,:,1] = G
    elif color == 3:
        imagen[:,:,0] = R
    elif color == 10:
        imagen[:,:,0:2] = cv.merge([R,G])
    elif color == 20:
        imagen[:,:,1:3] = cv.merge([G,B])
    elif color == 30:
        imagen[:,:,[0,2]] = cv.merge([R,B])
    else:
        print("Ingrese 1, 2, 3, 10, 20 o 30") 

    plt.imshow(imagen)
    
```

```python
canales("kirby.jpeg", 10)
```

#### Problema 2

```python
def original(azul,verde,rojo):
    azul = cv.imread(azul)
    verde = cv.imread(verde)
    rojo = cv.imread(rojo)

    # Asignacion de canales
    B = azul[:,:,0]
    G = verde[:,:,1]
    R = rojo[:,:,2]

    imagen = cv.merge([R,G,B])
    plt.imshow(imagen)
```

```python
original("imagen3/perro_salida_gray_azul.jpg", "imagen3/perro_salida_gray_verde.jpg", "imagen3/perro_salida_gray_rojo.jpg")
```

#### Problema 3

```python
def separacion(img):
    imagen = cv.imread(img)
    B,G,R = cv.split(imagen)
    
    cv.imwrite("canalRojo.jpg", R)
    RTemp = cv.imread("canalRojo.jpg", cv.COLOR_BGR2RGB)
    plt.imshow(RTemp, cmap="gray")
    plt.show()

    cv.imwrite("canalVerde.jpg", G)
    VTemp = cv.imread("canalVerde.jpg", cv.COLOR_BGR2RGB)
    plt.imshow(VTemp, cmap="gray")
    plt.show()

    cv.imwrite("canalAzul.jpg", B)
    BTemp = cv.imread("canalAzul.jpg", cv.COLOR_BGR2RGB)
    plt.imshow(BTemp, cmap="gray")
    plt.show()

    og = cv.merge([RTemp,VTemp,BTemp])
    plt.imshow(og)
```

```python
separacion("kirbo.jpg")
```

#### Problema 4

```python
def histogramas(img):
    img = cv.imread(img)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()

    R,G,B = cv.split(img)
    gris = (B+G+R)/3

    plt.hist(R.ravel(), bins=range(0,256,2), color='red')
    plt.axvline(np.mean(R), color='black', linestyle='dashed', linewidth=2)
    plt.title('Canal Rojo')
    plt.show()

    plt.hist(G.ravel(), bins=range(0,256,2), color='green')
    plt.axvline(np.mean(G), color='black', linestyle='dashed', linewidth=2)
    plt.title('Canal Verde')
    plt.show()

    plt.hist(B.ravel(), bins=range(0,256,2), color='blue')
    plt.axvline(np.mean(B), color='black', linestyle='dashed', linewidth=2)
    plt.title('Canal Azul')
    plt.show()

    plt.hist(gris.ravel(), bins=256, color='gray')
    plt.axvline(np.mean(gris), color='black', linestyle='dashed', linewidth=2)
    plt.title('Escala de grises')
    plt.show()
```

```python
histogramas("hokusai.png")
```

#### Problema 5

¿En qué consiste el espacio de color HSV y como se mapean colores a dicho especio?

El espacio de color HSV se refiere a que define un color en términos del *Hue*, o matiz, *Saturation*, o saturación, y un *Value*, o valor. El resultado de estas combinaciones son ruedas de colores que forman un cono. El tono (*Hue*) es representado como un grado entre 0° y 360°. 

La saturación, que se refiere a la pureza del color, se representa por valores entre 0 y 1. Estos valores se refieren a la distancia desde el centro de la rueda de color, mientras más cerca del centro, menos saturado será el color y viceversa. 

Y un valor asociado a su brillo, también entre 0 y 1. Este valor se refiere a la distancia vertical desde el centro a la rueda de color, mientras más cerca al punto del centro, el color será más oscuro y viceversa.
